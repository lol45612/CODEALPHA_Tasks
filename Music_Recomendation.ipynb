```python
import shutil

# Extract the zip file to the current working directory
shutil.unpack_archive(zip_file_path, '.')

# Load the extracted CSV file
df = pd.read_csv('Music Info.csv')
```

```python
print(df.head(10))
```

```python
# Convert 'Timestamp' to datetime
df['duration_ms'] = pd.to_datetime(df['duration_ms'])

# Extract relevant features
df['month'] = df['duration_ms'].dt.month
df['day'] = df['duration_ms'].dt.day
df['hour'] = df['duration_ms'].dt.hour

# Create a label for repeated plays within a month
df['repeated_play'] = df.duplicated(subset=['name', 'artist', 'year'], keep=False).astype(int)

# Display the first few rows to verify
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Define features and target
features = ['month', 'day', 'hour']
target = 'repeated_play'

# Split the data
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Train a Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
# Use the trained model to make predictions on new data
new_data = pd.DataFrame({
    'month': [5],
    'day': [15],
    'hour': [14]
})

prediction = model.predict(new_data)
print(f'Prediction for new data: {prediction}')
```
